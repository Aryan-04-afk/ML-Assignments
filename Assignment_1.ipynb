**Q3: Based on Searching and Sorting
(a) For the array: array = i. Sorted array
ii. Indices of sorted array
iii. 4 smallest elements
iv. 5 largest elements
(b) For the array: array = i. Integer elements only
ii. Float elements only
np.array([10, 52, 62, 16, 16, 54, 453]), find
np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0]), find**

```python
#3
import numpy as np

# (a)
array = np.array([10, 52, 62, 16, 16, 54, 453])

sorted_array = np.sort(array)
sorted_indices = np.argsort(array)
four_smallest = np.sort(array)[:4]
five_largest = np.sort(array)[-5:]

print("Sorted array:", sorted_array)
print("Indices of sorted array:", sorted_indices)
print("4 smallest elements:", four_smallest)
print("5 largest elements:", five_largest)

# (b)
array_float = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])

int_elements = array_float[array_float % 1 == 0]
float_elements = array_float[array_float % 1 != 0]

print("\nInteger elements only:", int_elements)
print("Float elements only:", float_elements)
```

 **Q4:
(a) Write a function named img_to_array(path) that reads an image from a specified path and
save it as text file on local machine? (Note: use separate cases for RGB and Grey Scale
images)
(b) Load the saved file into jupyter notebook?**

```python

from google.colab import files

uploaded = files.upload()  # Upload your image file here
# After upload, the filename(s) will be keys in `uploaded`


import cv2
import numpy as np
import os

def img_to_array(path):
    img = cv2.imread(path, cv2.IMREAD_UNCHANGED)
    if img is None:
        print("Error: Image not found or unable to read.")
        return

    if len(img.shape) == 2:
        print("Image is grayscale.")
        txt_filename = os.path.splitext(path)[0] + "_grayscale.txt"
        np.savetxt(txt_filename, img, fmt='%d')
        print(f"Grayscale image saved as text file: {txt_filename}")

    elif len(img.shape) == 3 and img.shape[2] == 3:
        print("Image is color (BGR).")
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        txt_filename = os.path.splitext(path)[0] + "_rgb.txt"
        reshaped = img_rgb.reshape(-1, 3)
        np.savetxt(txt_filename, reshaped, fmt='%d', header='R G B', comments='')
        print(f"RGB image saved as text file: {txt_filename}")
    else:
        print("Unsupported image format or number of channels.")

```
